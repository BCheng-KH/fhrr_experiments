```python
import numpy as np
import fhrr_driver_torch as fd
import torch
from matplotlib import pyplot as plt
import tqdm
import random
```

```python
SSP_space = fd.fhrr_space(10000)
base_vec = SSP_space.init_ones_vec()
num_vec = SSP_space.init_normal_avec(0.1*torch.pi)
```

```python
num = 10
num_ssp = base_vec*SSP_space.ssp_phase(num_vec, num)
```

```python
sample_range = [-10, 30]
step = 0.01
sample_nums = np.arange(sample_range[0], sample_range[1], step)
```

```python
sims = []
for i in range(len(sample_nums)):
    sample_ssp = base_vec*SSP_space.ssp_phase(num_vec, sample_nums[i])
    sims.append(SSP_space.similarity_C(num_ssp, sample_ssp))
sims = np.array(sims).flatten()
```

```python
plt.plot(np.real(sims))
```

```python
plt.plot(np.real(sims*np.exp(1j*np.pi/2)))
```

```python
plt.plot(np.real(sims*np.conj(sims)))
```

```python
plt.plot(np.real(sims))
plt.plot(np.real(sims*np.exp(1j*np.pi/2)))
```
