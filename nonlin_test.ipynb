```python
import numpy as np
import fhrr_driver_torch as fd
import torch
from matplotlib import pyplot as plt
import tqdm
import random
```

```python
vspace = fd.fhrr_space(10000)
```

```python
base = vspace.init_random_vec()
ssp = vspace.init_normal_avec(0.5)
def get_ssp(x):
    return vspace.bind(base, vspace.ssp_phase(ssp, x))
```

contraction test

```python
A = get_ssp(2)
B = get_ssp(-2)
targ_A = get_ssp(50)
targ_B = get_ssp(-50)
print(vspace.similarity_R(A, B), vspace.similarity_R(targ_A, targ_B))
```

```python
map1_in = fd.block_learning_map(vspace, vspace, 100, (100, 100))
map1_h = fd.block_learning_map(vspace, vspace, 100, (100, 100))
map1_out = fd.block_learning_map(vspace, vspace, 100, (100, 100))
for _ in range(1000):
    if random.choice([0, 1]) == 0:
        v = A
        targ_v = targ_A
    else:
        v = B
        targ_v = targ_B
    h1 = vspace.init_random_vec()
    h2 = vspace.init_random_vec()
    map1_in.learn(v, h1)
    map1_h.learn(h1,h2)
    map1_out.learn(h2, targ_v)
```

```python
print(vspace.similarity_R(map1_out.forwards(map1_h.forwards(map1_in.forwards(A))), map1_out.forwards(map1_h.forwards(map1_in.forwards(B)))))
```
