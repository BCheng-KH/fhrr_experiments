```python
import numpy as np
import fhrr_driver_torch as fd
import torch
from matplotlib import pyplot as plt
import tqdm
import random
```

```python
SSP_space = fd.fhrr_space(10000)
base_vec = SSP_space.init_ones_vec()
num_vec = SSP_space.init_normal_avec()
```

```python
n = 100
bundle = SSP_space.init_zeros_vec()
bundle_2 = SSP_space.init_zeros_vec()
for i in range(1, n+1):
    bundle += SSP_space.ssp_phase(num_vec, i)
    bundle_2 += SSP_space.init_random_vec()
bundle /= n**(1/2)
bundle_2 /= n**(1/2)
#bundle = SSP_space.normalize(bundle)
```

```python
r_vals = []
c_vals = []
for i in range(10000):
    r_vals.append(np.real(bundle[i,0]))
    c_vals.append(np.real(bundle[i,0]*np.exp(1j*-np.pi/2)))
```

```python
plt.scatter(r_vals, c_vals)
```

```python
print(np.mean(r_vals), np.mean(c_vals))
```

```python
r_vals2 = []
c_vals2 = []
for i in range(10000):
    r_vals2.append(np.real(bundle_2[i,0]))
    c_vals2.append(np.real(bundle_2[i,0]*np.exp(1j*-np.pi/2)))
plt.scatter(r_vals2, c_vals2, c='b')
plt.scatter(r_vals, c_vals, c='r')
print(np.mean(r_vals2), np.mean(c_vals2))
```
